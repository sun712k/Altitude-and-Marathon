# A. Introduction

As an avid fan of athletic events and a dreamer of becoming a runner myself, I have often heard claims that East Africans dominate distance running due to genetic, geographical, or socioeconomic factors. Among these, the geographical explanation—particularly high altitude—has always seemed the most plausible to me. Still, I wonder if that conclusion might be too simplistic. After all, there are other countries with even higher average altitudes, such as Switzerland, Nepal, or Peru, that do not produce the same number of elite runners. Moreover, many high-altitude countries also have low income levels, suggesting that socioeconomic factors could play an intertwined role. It is also worth noting that Kenya and Ethiopia, the two nations most frequently cited in discussions of distance-running dominance, are only two among the 18 countries classified as Eastern Africa according to the UN geoscheme.

To explore this further, I began by searching for general information on the topic and found the following:

*Google AI Search Summary:*

Yes, runners from high-altitude regions—particularly Kenya and Ethiopia—are disproportionately represented among the world’s top marathon finishers. While genetics may play a minor role, most research attributes their success to a combination of factors:

1. **High altitude** provides a physiological advantage by increasing red blood cell count, improving oxygen delivery, and creating optimal conditions for endurance training.
2. **An active lifestyle from a young age** develops strong aerobic capacity, as running is often a routine means of transportation.
3. **Motivation and social factors** such as the pursuit of economic opportunity and a strong running culture further reinforce excellence.

To gain a deeper understanding, I next reviewed scientific studies on this topic. Research on East African distance runners—particularly those from Kenya and Ethiopia—has been extensive, focusing on identifying the factors that contribute to their dominance in long-distance events. One well-known paper, *“Kenyan and Ethiopian Distance Runners: What Makes Them So Good?”*, discusses genetic, physiological, and socioeconomic explanations for their success.

> “In general, it appears that Kenyan and Ethiopian distance-running success is not based on a unique genetic or physiological characteristic. Rather, it appears to be the result of favorable somatotypical characteristics lending to exceptional biomechanical and metabolic economy/efficiency; chronic exposure to altitude in combination with moderate-volume, high-intensity training (live high + train high), and a strong psychological motivation to succeed athletically for the purpose of economic and social advancement.”

In terms of physiological factors, the authors found no distinctive advantage in maximal oxygen uptake or hematological profile. Regarding skeletal muscle–fiber composition and oxidative-enzyme profile, Kenyan runners showed no difference in type I muscle-fiber percentage or citrate synthase activity compared with Scandinavian runners, though they exhibited higher HADH activity, which may aid long-distance performance. However, this cannot explain success in shorter races such as the 1500 m or 5000 m. Comparable data for Ethiopian runners are currently unavailable.

Regarding somatotypical characteristics, the study notes that Kenyan and Ethiopian runners differ in body type—Kenyan runners tend to be more ectomorphic, while Ethiopians are relatively more mesomorphic. Yet, the authors caution that since biomechanical and metabolic efficiency in Ethiopians has not been directly measured, attributing East African success solely to somatotype remains premature.

When addressing the role of **living and training at high altitude**, the authors highlight that

> “a shared characteristic of the Kenyan and Ethiopian distance runners is the fact that most of them were born and raised at an elevation of approximately 2000 to 2500 m in eastern Africa.”

The study also reports quantified differences in training volume at vLT and vVO₂max between elite and international elite Kenyan athletes—levels that even elite runners struggle to sustain. Beyond these findings, however, claims about high-altitude effects on performance are largely observational or speculative. Nonetheless, the authors emphasize the importance of continued research, stating that

> “Regardless, it seems logical to assume that chronic altitude residence and moderate-volume, high-intensity altitude training may contribute in part to the exceptional performance of the Kenyan and Ethiopian distance runners and should not be disregarded.”

These findings raised several questions for me:
### Research Questions and Rationale

1. **Are many elite runners from Ethiopia and Kenya actually born in high-altitude regions?**

2. **What about runners from other countries with even higher altitudes?**

3. **Is there data on runners’ childhood regions to examine the validity of the “active lifestyle from a young age” claim?**
   Information on birthplace altitude would help assess the idea that East African runners “live high” — that is, spend their formative years in high-altitude environments that naturally enhance aerobic capacity.

4. **Is there data on runners’ main training locations?**
   Knowing where athletes train would address the “train high” component of the hypothesis. Since country-level average altitude may approximate the general training environment, it could serve as a preliminary indicator when detailed training-location data are unavailable.

5. **How do socioeconomic factors interact with performance?**
   It remains unclear whether athletes from low-income countries perform better due to strong motivation for economic advancement, or whether runners from high-income countries gain an advantage from superior financial support, facilities, and sports infrastructure.

6. **(Main Research Question 1)** ***Is country altitude associated with long-distance running performance?***
   Although national average altitude may not perfectly represent where runners live or train, it can still reveal partial effects of related covariates such as income, population size, and infrastructure — factors that shape sports culture and opportunities. However, an association observed at the country level does not necessarily imply a causal relationship at the individual level (an ecological fallacy).
   
7. **(Main Research Question 2)** ***Does altitude at the runner’s birth region predict performance within countries?***
   Since many researchers hypothesize that regional altitude influences performance through physiological mechanisms, the second stage of this study will analyze within-country variation—examining whether runners born in higher-altitude regions perform better after controlling for country-level factors. This approach helps isolate the potential physiological effect of altitude while minimizing confounding influences such as genetics and socioeconomic conditions.
   
   
   *If altitude is truly a causal driver of endurance performance (e.g., through oxygen adaptation or VO₂ max improvements), then the relationship should appear both *across* and *within* countries. Conversely, if the apparent country-level association is instead driven by cultural, genetic, or selection effects, the relationship may disappear when examined within countries.*

**Caveat:** The altitude of a runner’s birth region may not accurately reflect where they spent most of their life or training—factors that are more relevant for developing any long-term physiological adaptations. Similarly, a country’s average altitude may not represent the typical elevation of regions where most runners live or train. Nevertheless, using national average altitude provides a reasonable starting point to examine the broader claim that runners from higher-altitude countries tend to perform better.
